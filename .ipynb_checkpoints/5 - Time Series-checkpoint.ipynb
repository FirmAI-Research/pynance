# Prices

```python
prices = pd.DataFrame(yf.download(ticker, '2021-01-01', cal.today(), progress=False)['Adj Close'])
```

```python
plt.rcParams["figure.figsize"] = (20,7)
ts = prices.copy().reset_index()
x = ts['Date']
y = ts['Adj Close']
sns.lineplot(x = 'Date', y = 'Adj Close', data = ts)
```

```python
prices.plot(kind='kde')
```

```python
ts = TimeSeries(data=prices, column='Adj Close')
```

```python
ts.check_stationarity()
```

```python
ts.test_stationarity()
```

```python
ts.decomposition(model='multiplicative')
```

```python
# log trasnform non stationary data

ts.remove_nonstationary_trend()
```

## Arima

```python
a  =  Arima(df = pd.DataFrame(ts.train_data), col = 'Adj Close', order = (5, 1, 2))
```

```python
a.plot_correlogram()
```

```python
plt.rcParams["figure.figsize"] = (20,7)
a.arima_diagnostics(a.model_fit.resid, 40)
```
